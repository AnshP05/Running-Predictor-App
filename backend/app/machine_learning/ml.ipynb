```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("Ansh's RaceTime Predictor (Responses).xlsx")
```

```python
df[["1600m(or Mile) Time", "5k Time"]].head(10)
df.head()
```

```python
pair_df = df[["1600m(or Mile) Time", "5k Time"]].copy()
pair_df = pair_df.replace(0, np.nan)
pair_df = pair_df.dropna(subset=["1600m(or Mile) Time", "5k Time"])
print(pair_df)
```

```python
# Function to parse data
def time_to_seconds(time_str):
    parts = str(time_str).split(':')
    if len(parts) == 2:
        minutes = int(parts[0])
        seconds = float(parts[1])
        return minutes * 60 + seconds
    elif len(parts) == 3:
        hours = int(parts[0])
        minutes = int(parts[1])
        seconds = float(parts[2])
        return hours * 3600 + minutes * 60 + seconds
```

```python
pair_df["1600m_seconds"] = pair_df["1600m(or Mile) Time"].apply(time_to_seconds)
pair_df["5k_seconds"] = pair_df["5k Time"].apply(time_to_seconds)
print(pair_df)
```

```python
X = pair_df[["5k_seconds"]]
Y = pair_df["1600m_seconds"]

print(X.head())
print()
print(Y.head())
print()
print(X.shape, Y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)

print("X_train shape", X_train.shape)
print("X_test shape", X_test.shape)
print("Y_train shape", Y_train.shape)
print("Y_test shape", Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
# Make the predictions
predictions = model.predict(X_test)
print(np.round(predictions[:5], 0).astype(int))
```

```python
from sklearn.metrics import mean_absolute_error

ml_mae = mean_absolute_error(Y_test, predictions)
print("ML_MAE:", ml_mae)
```

```python
def riegel_predict(time1, distance1, distance2):
    return time1 * (distance2 / distance1) ** 1.06
```

```python
riegel_predictions = X_test["5k_seconds"].apply(
    lambda t : riegel_predict(t, 5000, 1600)
)

riegel_mae = mean_absolute_error(Y_test, riegel_predictions)
print("Riegel MAE:", riegel_mae)
```
